## Univariate Analysis

```python
# Job Title Analysis
print("\nJob Title:\n", job_data['job_title'].value_counts())
sns.histplot(x='job_title', data=job_data)
plt.xticks(rotation=45, ha='right')
plt.title('Job Title')
plt.show()
```

```python
# Company Location Analysis
print("\nCompany Location:\n", job_data['company_location'].value_counts())
sns.countplot(x='company_location', data=job_data)
plt.xticks(rotation=45, ha='right')
plt.title('Company Location')
plt.show()
```

```python
# Salary Currency Analysis
print("\nSalary Currencys:\n", job_data['salary_currency'].value_counts())
sns.countplot(x='salary_currency', data=job_data)
plt.xticks(rotation=45, ha='right')
plt.title('Salary Currency')
plt.show()
```

## Bivariate Analysis

```python
# Salary Currency based on Company Location
print("\nSalary Currency based on Company Location:\n", job_data.groupby('salary_currency')['company_location'].value_counts())
sns.countplot(x='salary_currency', hue='company_location', data=job_data)
plt.title('Salary Currency based on Company Location')
plt.show()
```

```python
# Salary based on Job Title
print("\nSalary based on Job Title:\n", job_data.groupby('job_title')['salary_usd'].value_counts())
sns.countplot(x='job_title', hue='salary_usd', data=job_data)
plt.xticks(rotation=45)
plt.title('Salary based on Job Title')
plt.show()
```

## Heatmap of numerical features

```python
plt.figure(figsize=(12, 8))
numeric_features = job_data.select_dtypes(include=np.number)
sns.heatmap(numeric_features.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Feature Correlation Heatmap')
plt.show()
```
